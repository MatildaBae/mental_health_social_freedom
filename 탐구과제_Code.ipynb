```r
# 필요한 패키지 설치 및 로드
library(readr)

# CSV 파일 불러오기
free <- read_csv("hfi_cc_2022.csv")
mental <- read_csv("mental_illnesses_prevalence.csv")
```

```r
# NaN 값이 382개보다 많은 변수 제거
free_cleaned <- free [, nan_count <= 385] # 88% 이상 값 있는

# 결과 확인
free_cleaned
```

```r
names(free_cleaned)
```

pf_rol : Rule of law(법치주의)

pf_ss : security and safety

pf_movement : freedom of movement

pf_religion : religious freedom

pf_association : freedom to associate and assemble with peaceful individuals and organizations

pf_expression : freedom to expression

pf_identity : identity and relationships

ef_government : size of government

ef_legal : legal system and property rights

ef_money : sound(stable) money(안정적이고 신뢰할 수 있는 통화 시스템)

ef_trade : freedom to trade internationally

ef_regulation : regulation(사회적 규제)

hf_score : human freedom index

```r
library(dplyr)

# 주요 변수 추출, 2019년까지
free_df <- free_cleaned %>% filter(year<2020) %>%
           select(year, countries,
           pf_rol, pf_ss, pf_movement,
           pf_religion, pf_assembly, pf_expression, pf_identity,
           ef_government, ef_legal, ef_trade, ef_money,
           ef_regulation, hf_score)

free_df
```

```r
# 각 변수별 NaN 값 수 계산
nan_count <- colSums(is.na(free_df))

# 결과 출력
nan_count
```

```r
# 그룹화 및 NaN 값 카운트
nan_counts <- free_df %>%
  select(countries, ef_trade) %>%
  group_by(countries) %>%
  summarise(across(everything(), ~ sum(is.na(.))))

# 결측치 비율이 높은 변수들에 한해서
# 20년 중 결측치가 15년치보다 많은 나라들 추출
nan <- nan_counts %>%
    filter(ef_trade > 15)

# 결과 확인
print(nan)
```

```r
# 결측치가 많은 나라들 제거
free_df <- free_df %>% filter(!countries %in% c('Belarus', 'Comoros', 'Djibouti',
                              'Iraq', 'Somalia', 'Sudan'))
```

```r
# 다른 변수(pf_rol)로 한 번 더 진행
# 그룹화 및 NaN 값 카운트
nan_counts <- free_df %>%
  select(countries, pf_rol) %>%
  group_by(countries) %>%
  summarise(across(everything(), ~ sum(is.na(.))))

nan <- nan_counts %>%
    filter(pf_rol > 15)

# 결과 확인
print(nan)
```

## Random Forest Modeling

### 연도 자체는 제거해야?

```r
# 'countries' 열 제거
new_data <- new_data %>% select(-countries)

# 'year' 열 제거
new_data <- new_data %>% select(-year)

# 데이터 분할
set.seed(123)
split <- sample.split(new_data$depressive, SplitRatio = 0.8)
train_data <- subset(new_data, split == TRUE)
test_data <- subset(new_data, split == FALSE)

# 데이터 스케일링
preProc <- preProcess(train_data, method = c("center", "scale"))
train_data_scaled <- predict(preProc, train_data)
test_data_scaled <- predict(preProc, test_data)
```

```r
# 스케일링 되었는지 확인
head(train_data_scaled)
```

```r
# 교차 검증 설정
train_control <- trainControl(method = "cv", number = 10)

# 랜덤 포레스트 모델 학습
set.seed(123)
rf_model <- train(depressive ~ ., data = train_data_scaled, method = "rf", trControl = train_control)
```

```r
# 모델 요약
print(rf_model)
```

### 변수 별 중요도 출력

```r
# 변수별 기여도 출력
importance <- varImp(rf_model, scale = FALSE)

print(importance)
plot(importance)
```
